```python
import torch
import torchvision.models as models
import torch.nn as nn
import torchvision
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from torch.nn import  functional as F
import numpy as np
from skimage.segmentation import slic
import itertools
```

```python
model = models.alexnet(pretrained=True)
preprocess = transforms.Compose([transforms.Resize(256), transforms.CenterCrop(224), transforms.ToTensor(), transforms.Normalize(mean = (0.485, 0.456, 0.406), std=(0.229, 0.224, 0.223))])
```

```python
D = torchvision.datasets.ImageFolder('/Users/pawel/Desktop/panda',transform = preprocess)
dataloader = torch.utils.data.DataLoader(D)
```

```python
trans1=transforms.ToPILImage()
```

```python
init_pred = []
maps = []
```

```python
for img in dataloader:
    init_pred.append(torch.max(model(img[0]), dim=1))
    maps.append(transforms.ToTensor()(slic(trans1(img[0].squeeze()), 99)))
```

```python
plt.imshow(maps[0].squeeze())
```

```python
torch.max(model(D[0][0].unsqueeze(dim=0)),dim=1)# 388 is a panda
```

```python
confidence = []
i=0
# here, I take a mask I want to delete from image and feed imagae without that part through the model


while i <99:  # goes through 99 superpixels
        panda = D[0][0].clamp(0,1)
        maper = ((maps[0]==i) | (maps[0]==i+1) | (maps[0] == i+2)).squeeze()
        conf = model((panda*(~maper)).unsqueeze(dim=0)) # I hate this part, its ugly. 
        i+=3
        confidence.append(conf[0][388].item())
```

```python
# than I take top 10 worst performing images to see which superpixels had most impact
best_performers = sorted(enumerate(confidence), key = lambda x: x[1])[:10]
best_performers # 10 most important superpixlels
```

```python
best_performers_ids = {x[0]+b for x in best_performers for b in (0,1,2)}
```

```python
best_performers_ids
```

```python
# this goes against dry, tbd
confidence2 = []
for i in itertools.combinations(best_performers_ids,3):
    panda = D[0][0].clamp(0,1)
    maper = ((maps[0]==i[0]) | (maps[0]==i[1]) | (maps[0] == i[2])).squeeze()
    conf = model((panda*(maper)).unsqueeze(dim=0))
    confidence2.append((i,conf[0][388].item()))
```

```python
final_list = sorted(confidence2, key = lambda x: x[1], reverse = True)
final_list
```

```python
final_map = ((maps[0]==final_list[0][0][0]) | (maps[0]==final_list[0][0][1]) | (maps[0] == final_list[0][0][2])).squeeze()
final_image = (D[0][0]*(final_map)).unsqueeze(dim=0)
plt.imshow(final_image.squeeze().permute(1,2,0))
```
