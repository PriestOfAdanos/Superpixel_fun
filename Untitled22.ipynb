```python
import cv2
import numpy as np
from skimage.segmentation import slic
from matplotlib import pyplot as plt
from collections import Counter
```

```python
img = cv2.imread("/content/8983023-202200325")
```

```python
segments = slic(img, n_segments = 500, compactness = 10)
```

```python
output = np.copy(img)
```

```python
for segment_val in np.unique(segments):
    mask = np.zeros(img.shape[:2], dtype="uint8")
    mask[segments == segment_val] = 255
    color = cv2.mean(img, mask=mask)[:3]
    output[mask == 255] = color
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].axis("off")
ax[0].set_title("Original Image")

ax[1].imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
ax[1].axis("off")
ax[1].set_title("Segmented Image")

plt.show()
```

```python
colors = np.zeros((np.max(segments)+1, 3), dtype=np.float64)
for i in np.unique(segments):
    colors[i] = np.mean(img[segments == i], axis=0)

red = colors[:, 2]
green = colors[:, 1]
blue = colors[:, 0]

plt.figure(figsize=(16, 6))

plt.subplot(131)
n, bins, patches = plt.hist(red, bins=50, color='red')
plt.title("Red color distribution")
plt.xlabel("Color intensity")
plt.ylabel("Frequency")

plt.subplot(132)
n, bins, patches = plt.hist(green, bins=50, color='green')
plt.title("Green color distribution")
plt.xlabel("Color intensity")
plt.ylabel("Frequency")

plt.subplot(133)
n, bins, patches = plt.hist(blue, bins=50, color='blue')
plt.title("Blue color distribution")
plt.xlabel("Color intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
sizes = Counter(segments.flatten())
plt.figure(figsize=(10,5))
plt.bar(sizes.keys(), sizes.values())
plt.title("Size distribution of superpixels")
plt.xlabel("Superpixel label")
plt.ylabel("Size")
plt.show()
```

```python
centers = np.zeros((np.max(segments)+1, 2), dtype=np.float64)
for i in np.unique(segments):
    centers[i] = np.mean(np.nonzero(segments == i), axis=1)
plt.figure(figsize=(10,5))
plt.scatter(centers[:,1], centers[:,0])
plt.title("Spatial distribution of superpixels")
plt.gca().invert_yaxis()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
